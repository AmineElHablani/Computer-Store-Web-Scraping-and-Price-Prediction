```python
#Get the columns that have a significant difference with price 

columns=[x for x in data.columns if x!='price' and x!= "Destockage"]

significant_columns=[]

for col in columns:
    test=data[col].value_counts()[data[col].value_counts() > 20 ].keys()
    test=data[data[col].isin(test)]
    if test.shape[0]>40:
        #anova 
        result= pingouin.anova(data=test,
                               dv='price',
                               between=col)
        if result['p-unc'].item() < 0.01:
            significant_columns.append(col)

        
```

## Label Encoding

### Prepare data

```python
label=[]
```

```python
data_label_encoding=pd.DataFrame(data['price'])
```

```python
#Create New data using label encoding

for col in significant_columns:
    new_column=[]
    unique_elements=data[col].unique().tolist()
    
    for element in data[col]:
        for idx,unique_el in enumerate(unique_elements):
            if unique_el==element:
                new_column.append(idx)
    data_label_encoding[col]=new_column

                
```
